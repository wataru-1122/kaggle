```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
data=pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
target=boston.target
```

```python
plt.hist(target,bins=50)
plt.show
```

```python
for i in data.columns:
    plt.hist(data[i],bins=30)
    plt.title(i)
    plt.show()
```

```python
data.columns
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val=train_test_split(data,target,test_size=0.2,random_state=0)

lr.fit(X_train,y_train)

y_pred=lr.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_val,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest=RandomForestRegressor()
forest.fit(X_train,y_train)
y_pred_forest=forest.predict(X_val)
```

```python
mean_squared_error(y_val,y_pred_forest)
```

```python
from sklearn.model_selection import GridSearchCV
forest=RandomForestRegressor()

param_dit={'n_estimators':i for i in range(100,200)}
gs=GridSearchCV(estimator=forest,param_grid=param_dit,scoring='r2',cv=5,n_jobs=-1)
gs.fit(data,target)
```

```python
gs.best_params_
```

```python
best_model=gs.best_estimator_
```

```python
best_model.fit(X_train,y_train)
y_pred=best_model.predict(X_val)
```

```python
mean_squared_error(y_val,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
forest=RandomForestRegressor(n_estimators=107)

param_dit={'max_depth':[i for i in range(5,20)]}
gs=GridSearchCV(estimator=forest,param_grid=param_dit,scoring='r2',cv=5,n_jobs=-1)
gs.fit(data,target)
```

```python
gs.best_params_
```

```python
best_forest=RandomForestRegressor(n_estimators=107,max_depth=12)
best_forest.fit(X_train,y_train)
y_pred_best=best_forest.predict(X_val)
```

```python
mean_squared_error(y_val,y_pred_best)
```
